```python
def find_task(spec, task_id):
    for hit in spec['hits']:
        sections = hit['tasks'][-2]['tasks'][-3:-1]
        for sec in sections:
            for task_pair in sec[-4]['tasks']:
                # print(task)
                if task_id in task_pair[1]['id']:
                    print(f'{hit["id"]} -> {task_pair[1]["id"]}')
            
```
